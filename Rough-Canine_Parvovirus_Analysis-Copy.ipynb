1. Prevalence of Canine Parvovirus infection
Use the parvo_data_reshaped array to calculate the prevalence of Canine Parvovirus infection among dogs in Akure South over the study period.


```python
# Calculate the prevalence of Canine Parvovirus infection
total_infected_days = (parvo_data_reshaped > 0).sum(dim='time')
total_days = parvo_data_reshaped['time'].size

# Calculate prevalence as a percentage
prevalence = (total_infected_days / total_days) * 100

# Print or visualize the results
print(f"Prevalence of Canine Parvovirus infection: {prevalence.values.item():.2f}%")
```

```python
import pandas as pd
import numpy as np
import xarray as xr

# Your existing code

# Create a DataFrame for infected days by year
df_infected_days_by_year = pd.DataFrame({
    'year': parvo_data_reshaped_grouped_by_year['year'].values,
    'infected_days': parvo_data_reshaped_grouped_by_year.values
})

# Create a DataFrame for total days by year
df_total_days_by_year = pd.DataFrame({
    'year': total_days_by_year['year'].values,
    'number_of_days': total_days_by_year.values
})

# Merge the two DataFrames on the 'year' column
df_result = pd.merge(df_infected_days_by_year, df_total_days_by_year, on='year')

# Display the result
df_result
```

```python
# Create a DataArray with a single 'time' dimension
parvo_data = xr.DataArray(
    parvo['PARVO'].values,
    coords={'time': pd.to_datetime(parvo['YEARS'].astype(str) + '-' + parvo['DAYS'].astype(str), format='%Y-%j')},
    dims=['time']
)

# Reshape the array to (1, 132)
parvo_data_reshaped = parvo_data.expand_dims(dim='dummy').transpose('dummy', 'time')

# Optionally, remove the 'dummy' dimension
parvo_data_reshaped = parvo_data_reshaped.squeeze('dummy')
```

```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming `parvo_data_reshaped` and `temp_data` are your xarray DataArrays

# Create new DataArrays with names
parvo_data_reshaped_named = xr.DataArray(parvo_data_reshaped.values, dims=['time'], coords={'time': parvo_data_reshaped['time']}, name='parvo_data')
temp_data_named = xr.DataArray(temp_data['t2m'].values, dims=['time'], coords={'time': temp_data['time']}, name='temp_data')

# Merge parvo_data_reshaped_named with temp_data_named on the 'time' dimension
merged_data = xr.merge([parvo_data_reshaped_named, temp_data_named], join='inner')

# Convert the merged data to a Pandas DataFrame
df_merged = merged_data.to_dataframe().dropna()

# Calculate correlation between temperature and CPV infections
correlation = df_merged['parvo_data'].corr(df_merged['temp_data'])

# Display the correlation
print(f"Correlation between temperature and CPV infections: {correlation:.2f}")

# Plotting the relationship
plt.figure(figsize=(10, 6))
plt.scatter(df_merged['temp_data'], df_merged['parvo_data'], alpha=0.5)
plt.title('Correlation Between Temperature and CPV Infections')
plt.xlabel('Temperature')
plt.ylabel('CPV Infections')
plt.grid(True)

# Save the plot in the "figures" folder with a corresponding name
figure_name = os.path.join(figure_folder, 'temperature_cpv_correlation.png')
plt.savefig(figure_name)

# Show the plot
plt.show()

print(f"Plot saved as: {figure_name}")
```

```python
import xarray as xr
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming `parvo_data_reshaped` and `temp_data` are your xarray DataArrays

# Create new DataArrays with names
parvo_data_reshaped_named = xr.DataArray(parvo_data_reshaped.values, dims=['time'], coords={'time': parvo_data_reshaped['time']}, name='parvo_data')
temp_data_named = xr.DataArray(temp_data['t2m'].values, dims=['time'], coords={'time': temp_data['time']}, name='temp_data')

# Merge parvo_data_reshaped_named with temp_data_named on the 'time' dimension
merged_data = xr.merge([parvo_data_reshaped_named, temp_data_named], join='inner')

# Convert the merged data to a Pandas DataFrame
df_merged = merged_data.to_dataframe().dropna()

# Calculate correlation matrix
correlation_matrix = df_merged.corr()

# Plot correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap between Temperature and CPV Infections')

# Save the plot in the "figures" folder with a corresponding name
figure_name = os.path.join(figure_folder, 'temperature_cpv_correlation_heatmap.png')
plt.savefig(figure_name)

# Show the plot
plt.show()

print(f"Plot saved as: {figure_name}")
```
